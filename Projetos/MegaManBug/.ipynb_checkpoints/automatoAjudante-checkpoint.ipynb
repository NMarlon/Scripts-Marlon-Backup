## Lógica
### Lógica negativa
- Lógica negativa, a contrapartida da lógica, a lógica que é verdadeira POR CAUSA que em outra perspectiva é FALSA. São coisas onde é possível saber o que é, pelo o que não é 
    - Ex.: 
        - É um número de 0 à 10, mas não é um número de 0 à 9.
            - é o próprio 10.
        - Se isto não é água, então não precisa pensar em variações de MAR, LAGO ou CHUVA pra descobrir o que é. 
        - Se isso não contém água, além das de cima, também não é garrafa ou copos, ou seja, a coisa nunca foi isso.
        
        
### Lógica mental em relação ao Tempo:
- Um delay no tempo externo não deve interferir numa lógica correta, afinal, é só esperar que vai dar certo, Á MENOS que haja algo que julgue pelo todo, então se máquina ainda não saber de tudo sobre algo (o que é normal à princípio) se julgar precipitadamente que aquilo não é real porque ainda não aconteceu, vai ocorrer falha e julgamento precipado, saindo do que deveria fazer que era só esperar e caindo numa falsa lógica, uma dicotomia da ansiedade e precipitação..
- Agora se o delay for interno pro externo, é como se fizesse algo agora, mas só saísse depois, isso pode e vai fazer com que algumas coisas que bot faça chegem após o momento da ação já haver passado, com isso, ela está limitada, ou age pela fé (sabe o que quer fazer/o que vai acontecer e faz antes pra compensar o delay) ou reconhece os novos limites.
